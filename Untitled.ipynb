```python
def process(pic,erode=None,dial=None,blur=None,threshold=0):
    if erode != None:
        pic = cv2.erode(pic, erode, iterations=1)
    if dial != None:
        pic = cv2.dial(pic, erode, iterations=1)
    if blur != None:
        pic = cv2.GaussianBlur(pic, blur, interations=1)
    if threshold > 0:
        pic = cv2.threshold(pic, threshold, 255, cv2.THRESH_BINARY)[1]
    return pic
```
